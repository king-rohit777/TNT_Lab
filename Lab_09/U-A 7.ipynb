### Find the set of columns which are not co related to each other.

```python
corr_mat = df.corr()
corr_mat
```

```python
for a, column_a in corr_mat.transpose().iterrows():
    for b, column_b in corr_mat.transpose().iterrows():
        if corr_mat[b][a]!=1.0:
            print(f"({a},{b})") 
```
